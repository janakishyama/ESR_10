```python
#import libraries
import numpy as np
import pandas as pd
import os
import re
from datetime import date
import sys

# to avoid warning during executions
import warnings    
warnings.filterwarnings("ignore")
```

```python
# load all the file names into a list (files in the pbr data folder)
os.getcwd()
_dir_path = '../../../dataExport/PBRdata'
raw_data = os.listdir(_dir_path)
raw_data
```

```python
# join file name with path - file name can be specified as list item based on required file name
filename=  os.path.join(_dir_path, raw_data[3])
filename
```

```python
trial_no = re.findall(r'\d+', filename)
trial_no = int(trial_no[0])
trial_no
```

```python
# load the selected file into the data frame.
pbr_rawDF = pd.read_csv(filename, sep=',')
pbr_rawDF.head(5)
```

```python
pbr_rawDF.insert(loc=0, column='trial_no', value=trial_no)
```
